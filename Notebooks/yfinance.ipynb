```python
import yfinance as yf
import pandas as pd
import numpy as np
import xlwings as xw
import datetime as dt
```

```python
def write_xl():
    """
    Args:
        tickers (list of str): 
    """
    wb = xw.Book.caller()
    sheet0 = wb.Sheets[0]
    
    ticker = sheet0["A2"]
    start_date = pd.to_datetime(sheet0["B2"], format=sheet0["D2"])
    end_date = pd.to_datetime(sheet0["C2"], format=sheet0["D2"])
    
    data = yf.download(ticker, start_date, end_date)
    wb.sheets[0].range("A5").value = data
```

```python
tickers = ["MSFT", "TSLA", "AAPL"]

# start_date = "01.01.2024"
# end_date = "01.02.2024"

start_date = "2024/01/01"
end_date = "2024/02/01"

write_xl(tickers, start_date, end_date, date_format="%Y/%m/%d")
```
